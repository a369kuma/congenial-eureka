```python
day_type = input("Is it a workday or weekend? ").lower().strip()
hours = float(input("Enter the total hours worked: "))

total_pay = 0
```

```python
if day_type == "workday":
    if hours <= 8:
        total_pay = hours * 15
    else:
        overtime_hours = hours - 8
        total_pay = (8 * 15) + (overtime_hours * 21)
else:
    total_pay = hours * 21
```

```python
print(f"The employee should be paid: ${total_pay:.2f}")
```
