```python
percent = float(input("Enter your percentage grade (0-100): "))
```

```python
if percent >= 90:
    gpa = 4.0
elif percent >= 85:
    gpa = 3.9
elif percent >= 80:
    gpa = 3.7
elif percent >= 77:
    gpa = 3.3
elif percent >= 73:
    gpa = 3.0
elif percent >= 70:
    gpa = 2.7
elif percent >= 67:
    gpa = 2.3
elif percent >= 63:
    gpa = 2.0
elif percent >= 60:
    gpa = 1.7
elif percent >= 50:
    gpa = 1.0
else:
    gpa = 0.0
```

```python
print(f"Your GPA for this course is: {gpa}")
```
