```python
import random

def run_quiz(filename):
    questions = []
    with open(filename, 'r') as f:
        for line in f:
            questions.append(line.strip().split('|'))

    total_avail = len(questions)
    num_to_test = int(input(f"How many questions (1-{total_avail})? "))

    test_set = random.sample(questions, num_to_test)
    results = []
    score = 0
    
    for q in test_set:
        print(f"\n{q[0]}")
        for i in range(1, 5):
            print(f"  {i-1}: {q[i]}")
        
        ans = int(input("Your answer (0-3): "))
        correct_idx = int(q[5])
        
        is_correct = (ans == correct_idx)
        if is_correct: score += 1
            
        results.append({
            'q': q[0], 
            'correct': is_correct, 
            'given': q[ans+1], 
            'right': q[correct_idx+1]
        })

    print(f"\n--- Score: {(score / num_to_test) * 100:.1f}% ---")
    
    for r in results:
        if not r['correct']:
            print(f"Incorrect: {r['q']}")
            print(f"  You said: {r['given']} | Correct: {r['right']}")

run_quiz('questions.txt')
```
