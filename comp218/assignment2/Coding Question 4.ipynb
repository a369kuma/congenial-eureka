```python
def Hanoi(n, source, target, temp):
    if n == 1:
        return [f"{source} -> {target}"]
    
    moves = []

    moves.extend(Hanoi(n - 1, source, temp, target))
    
    moves.append(f"{source} -> {target}")
    
    moves.extend(Hanoi(n - 1, temp, target, source))
    
    return moves
```
